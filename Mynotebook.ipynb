First of all, to consider only the houses built year 2000 and onwards, i created a new function that is a copy of the first one but includes only the houses where year built was 2000 and onwards.

```python
df_2 = df.loc[df['yr_built'] >= 2000].copy()
```

```python
df_2.head()
```

```python
df_2['yr_built'].value_counts()
```

```python
maxprice_year = df_2.groupby(df_2['yr_built'])['price'].idxmax()
most_expensive_houses = df_2.loc[maxprice_year, ['yr_built', 'price']]
print(most_expensive_houses)
```

```python
max_bedrooms = df_2.groupby(df_2['yr_built'])['bedrooms'].idxmax()
max_bedrooms_df = df_2.loc[max_bedrooms, ['id', 'yr_built','bedrooms']]
print(max_bedrooms_df)
```

```python
max_bathrooms = df_2.groupby(df_2['yr_built'])['bathrooms'].idxmax()
max_bathrooms_df = df_2.loc[max_bathrooms, ['id', 'yr_built','bathrooms']]
print(max_bathrooms_df)
```

```python
max_floors = df_2.groupby(df_2['yr_built'])['floors'].idxmax()
max_floors_df = df_2.loc[max_floors, ['id', 'yr_built','floors']]
print(max_floors_df)
```

```python
import math 
def calc_distance(coordinate1, coordinate2):
    lat1, long1  = coordinate1 
    lat2, long2 = coordinate2 
    distance = math.acos(math.sin(lat1)*math.sin(lat2)+math.cos(lat1)*math.cos(lat2)*math.cos(long2-long1))*6371
    return distance
```

**Now concerning the .at and .iat methods in pandas**

The .at and .iat methods are used to access specific elements in a dataframe using either their labels (.at method) or their index(.iat method)
They are similar to the loc and iloc but are faster and more efficient

Example:

```python
value_at = df.at[2, 'sqft_living']
print(value_at)
```

```python
value_iat = df.iat[2, 4]  ##row then column
print(value_iat)
```
